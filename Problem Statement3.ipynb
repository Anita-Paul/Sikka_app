```python
import pandas as pd
from sklearn.linear_model import LinearRegression

```

```python
# Load the data for October
data = pd.read_csv('Q3_Sikka data.csv', parse_dates=['Date'], index_col='Date')

data
```

```python
data.head()
```

```python
data.columns
```

```python
#Prepare the data
# Select the columns needed for training the model
X = data.loc['2022-10-01':'2022-10-31', ['Installs', 'DAU']]
y = data.loc['2022-10-01':'2022-10-31', 'Referrals']

# Print the first 5 rows of X and y
print(X.head())
print(y.head())
```

```python
#Train the model
from sklearn.linear_model import LinearRegression

# Create an instance of the linear regression model
lr_model = LinearRegression()

# Fit the model to the data
lr_model.fit(X, y)
```

```python
#Make predictions
# Select the columns needed for making predictions
X_pred = data.loc['2022-11-01':'2022-11-15', ['Installs', 'DAU']]

# Make predictions
y_pred = lr_model.predict(X_pred)

# Print the predictions
print(y_pred)
```
